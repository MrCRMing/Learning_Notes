```python
from sklearn.datasets import load_svmlight_file 
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
#读取数据并划分训练集合和验证集
X_train,y_train=load_svmlight_file("a9a_data")
X_val,y_val=load_svmlight_file("a9a_test_data",n_features=123)
X_train=X_train.toarray()
X_val=X_val.toarray()
y_train=y_train.reshape(y_train.shape[0],1)
y_val=y_val.reshape(y_val.shape[0],1)
print("X_train shape",X_train.shape)
print("X_val shape",X_val.shape)
print("y_train shape",y_train.shape)
print("y_val shape",y_val.shape)
```

```python
#初始化模型参数
W=np.zeros((X_train.shape[1],1))
b=0
print('W_shape',W.shape)
```

```python
#定义loss函数
def Loss(W,b,X,y,C):
    n=X.shape[0]
    temp1=1/2 * np.sqrt(np.sum(np.power(W,2).shape,axis=0)) 
    temp2=X.dot(W)+b
    temp3=1-y*(temp2)
    temp4=np.maximum(temp3,0)
    loss=  temp1  + C/n * sum(temp4)
    return loss
```

```python
#测试loss函数
loss=Loss(W,b,X_train[:200],y_train[0:200],0.1)
print(loss)
```

```python
#定义预测函数，
def predict(W,b,X):
    m=X.shape[0]
    y=X.dot(W)+b
    Y_prediction=np.zeros((m,1))
    for i in range(m):
        if y[i,0]>0:
            Y_prediction[i,0]=1
        else:
            Y_prediction[i,0]=-1
    return Y_prediction
```

```python
#随机批量梯度下降
def gradientDescent(W,b,X_train,y_train,X_val,y_val,alpha,C,method,epoch=1000,random_seed=None,batch_size=None):
    """
    计算导数,并更新W,同时计算训练损失和验证损失
    """
    train_loss=np.zeros(epoch)
    val_loss=np.zeros(epoch)
    m=X_train.shape[0]
    if method=="MBGD":#小批量随机梯度下降 
        np.random.seed(random_seed)
        rand_arr=np.arange(X_train.shape[0])
        for i in range(epoch):
            train_loss[i]=Loss(W,b,X_train,y_train,C)
            val_loss[i]=Loss(W,b,X_val,y_val,C)
            np.random.shuffle(rand_arr)
            batch_size=int(batch_size)
            index=rand_arr[0:batch_size]
            X_train_batch=X_train[index]
            y_train_batch=y_train[index]
            gW=np.zeros(W.shape)
            gb=0
            #更新参数W,b
            for i in range(X_train_batch.shape[0]):
                if y_train_batch[i]*(X_train_batch[i].dot(W)+b)<=1:
                    gW=gW+(-y_train_batch[i]*X_train_batch[i]).reshape(-1,1)
                    gb=gb+(-y_train_batch[i])


            DW=W+C/batch_size*gW
            Db=C/batch_size*gb

            W=W-alpha*DW
            b=b-alpha*Db

        return W,b,train_loss,val_loss
    elif method=="SGD":#随机梯度下降
        np.random.seed(random_seed)
        rand_arr=np.arange(X_train.shape[0])
        for i in range(epoch):
            train_loss[i]=Loss(W,b,X_train,y_train,C)
            val_loss[i]=Loss(W,b,X_val,y_val,C)
            np.random.shuffle(rand_arr)
            index=rand_arr[0:1]
            X_train_one=X_train[index]
            y_train_one=y_train[index]
            gW=np.zeros(W.shape)
            gb=0
            #更新参数W,b
            if y_train_one*(X_train_one.dot(W)+b)<=1:
                gW=gW+(-y_train_one*X_train_one).reshape(-1,1)
                gb=gb+(-y_train_one)


            DW=W+C*gW
            Db=C*gb

            W=W-alpha*DW
            b=b-alpha*Db

        return W,b,train_loss,val_loss
    else:#批量梯度下降
        for i in range(epoch):
            train_loss[i]=Loss(W,b,X_train,y_train,C)
            val_loss[i]=Loss(W,b,X_val,y_val,C)
            gW=np.zeros(W.shape)
            gb=0
            m=X_train.shape[0]
            #更新参数W,b
            for i in range(m):
                if y_train[i]*(X_train[i].dot(W)+b)<=1:
                    gW=gW+(-y_train[i]*X_train[i]).reshape(-1,1)
                    gb=gb+(-y_train[i])


            DW=W+C/m*gW
            Db=C/m*gb

            W=W-alpha*DW
            b=b-alpha*Db

        return W,b,train_loss,val_loss
```

```python
#使用随机批量梯度下降
epoch=3000
alpha=1
C=0.001
new_W,new_b,train_loss,val_loss=gradientDescent(W,b,X_train,y_train,X_val,y_val,alpha,C,method="MBGD",epoch=epoch,random_seed=2019,batch_size=300)
```

```python
#画出学习曲线
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(np.arange(epoch), train_loss, 'r',label='train')
ax.plot(np.arange(epoch), val_loss, color='green', label='val')
plt.legend() 
ax.set_xlabel('Interations')
ax.set_ylabel('Loss')
ax.set_title('Error vs. Training Epoch in MBGD')
plt.show()
```

```python
#计算在验证集的准确率
Y_prediction=predict(new_W,new_b,X_val)
correct = [1 if a==b else 0 for (a, b) in zip(Y_prediction, y_val)]
accuracy = sum(correct) / len(correct)
print("正确率：",accuracy)
```
